```python
import networkx as nx
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

bama = pd.read_csv("https://raw.githubusercontent.com/acgowda/cfb-recruit-net/main/data/2022/alabama_offers.csv")
bama_new = pd.DataFrame(data=bama['school'])
bama_new['university'] = 'Alabama'
#bama_new = bama_new.drop_duplicates(subset=['school'])

g = pd.read_csv("https://raw.githubusercontent.com/acgowda/cfb-recruit-net/main/data/2022/georgia_offers.csv")
g_new = pd.DataFrame(data=g['school'])
g_new['university'] = 'Georgia'
#g_new = g_new.drop_duplicates(subset=['school'])

ucla = pd.read_csv("https://raw.githubusercontent.com/acgowda/cfb-recruit-net/main/data/2022/ucla_offers.csv")
ucla_new = pd.DataFrame(data=ucla['school'])
ucla_new['university'] = 'UCLA'
#ucla_new = ucla_new.drop_duplicates(subset=['school'])

usc = pd.read_csv("https://raw.githubusercontent.com/acgowda/cfb-recruit-net/main/data/2022/usc_offers.csv")
usc_new = pd.DataFrame(data=usc['school'])
usc_new['university'] = 'USC'
#usc_new = usc_new.drop_duplicates(subset=['school'])


res = pd.concat([bama_new, g_new, ucla_new, usc_new])

schools = res['school'].values.tolist()

G = nx.from_pandas_edgelist(res,'school', 'university', create_using=nx.MultiGraph())
color_map = ['red' if node in schools else 'blue' for node in G]  
pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels = True, node_color = color_map, connectionstyle= 'arc3, rad = 0.1')
```
